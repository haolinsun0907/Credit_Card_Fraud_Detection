# Exploratory_Data_Analysis

Start by importing the necessary libraries and load the dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df = pd.read_csv("creditcard.csv")
#Peek into the first few rows of the dataset:
df.head()
#Get dataset summary:
df.describe()
```

```python
#Check for missing values:
df.isnull().sum()
```

Understanding Target Variable:

```python
#Check the distribution of the 'Class' variable to understand the imbalance:
sns.countplot(x='Class', data=df)
plt.title('Class Distribution (0: No Fraud | 1: Fraud)')
#Calculate percentage of fraud cases:
fraud_percentage = (df['Class'].sum() / df['Class'].count()) * 100
print(f"Fraudulent Transactions: {fraud_percentage:.2f}%")
```

Time and Amount Analysis

```python
#Plot distribution of the 'Time' fewture
plt.figure(figsize = (10,6))
sns.distplot(df['Time'],bins=100)
plt.title('Transaction Time Distribution')
plt.xlabel('Time (in seconds)')
```

```python
#Plot distribution of 'Amount' feature for both fraud and non-fraud transactions:
plt.figure(figsize=(10, 6))
sns.distplot(df[df['Class'] == 0]['Amount'], bins=100, color='green', label='Legitimate')
sns.distplot(df[df['Class'] == 1]['Amount'], bins=100, color='red', label='Fraudulent')
plt.title('Transaction Amount Distribution')
plt.xlabel('Amount ($)')
plt.legend()
```

Understand PCA Features:

```python
#Plot a histogram of all PCA transformed features to check their distribution:
plt.figure(figsize=(20, 15))
for i, col in enumerate(df.columns[1:29], 1):
    plt.subplot(7, 4, i)
    sns.histplot(df[col], bins=50, color='blue')
    plt.title(col)
    plt.tight_layout()
```

Correlation Analysis:

```python
#Plot a heatmap to check for correlations between features, especially with the 'Class' variable:
correlation_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
```

Boxplots for Anomaly Detection:
For features that seem to have a correlation with the 'Class' variable, plot boxplots to visually inspect for differences in their distribution across classes:

```python
# For feature V17
sns.boxplot(x='Class', y='V17', data=df)
plt.title('V17 vs Class')
```
