# Credit Card Fraud Detection

Start by importing the necessary libraries and load the dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df = pd.read_csv("creditcard.csv")
#Peek into the first few rows of the dataset:
df.head()
#Get dataset summary:
df.describe()
```

```python
#Check for missing values:
df.isnull().sum()
```

Understanding Target Variable:

```python
#Check the distribution of the 'Class' variable to understand the imbalance:
sns.countplot(x='Class', data=df)
plt.title('Class Distribution (0: No Fraud | 1: Fraud)')
#Calculate percentage of fraud cases:
fraud_percentage = (df['Class'].sum() / df['Class'].count()) * 100
print(f"Fraudulent Transactions: {fraud_percentage:.2f}%")
```
